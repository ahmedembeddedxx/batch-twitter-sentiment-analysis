### Necessary Libraries

```python
from wordcloud import WordCloud
import numpy as np
from pyspark.sql import SparkSession
from pyspark.sql.functions import regexp_replace, col, lower, expr, udf, max, min
from pyspark.ml.feature import HashingTF, IDF
from pyspark.ml.feature import CountVectorizer
from pyspark.ml import Pipeline
from pyspark.ml.feature import StopWordsRemover
from pyspark.sql.types import DoubleType, StringType
from pyspark.ml.feature import StringIndexer
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import MulticlassMetrics
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from pyspark.ml.feature import Tokenizer
```

### Spark Session

```python
spark = SparkSession.builder \
    .appName("HadoopDataFetch") \
    .config("spark.hadoop.fs.defaultFS", "hdfs://127.0.0.1:9000") \
    .getOrCreate()
```

### Hadoop Cluster to Spark DataFrame

```python
data_path = "/user/ahmadabdullah/twitter/sentiment.csv"

df = spark.read.format("csv").option("header", "true").option("encoding", "utf-8").load(data_path)

df.columns
```

```python
columns_to_keep = ['sentiment', 'sentiment_confidence', 'text']

df = df[columns_to_keep]
df.show(10)
```

```python
df = df.withColumn("sentiment_confidence", df["sentiment_confidence"].cast(DoubleType()))
```

```python
df.printSchema()
```

```python
df.count()
```

```python
min_max_confidence = df.selectExpr("min(sentiment_confidence) as min_confidence", "max(sentiment_confidence) as max_confidence").first()

min_confidence = min_max_confidence.min_confidence
max_confidence = min_max_confidence.max_confidence

print(f"Min. Confidence: {min_confidence}")
```

### Dropping Nulls, Retweets & Duplicate

```python
df = df.na.drop()
df = df.dropDuplicates()
```

```python
df = df.filter(~df['text'].startswith('RT'))
df = df.filter(~df['text'].startswith('"RT'))
df = df.filter((df['sentiment_confidence'] > 0.5) & (df['sentiment_confidence'] <= 1))

df.show(5)
```

```python
def remove_at_word(text):
    words = text.split()
    cleaned_words = [word for word in words if '@' not in word]
    return ' '.join(cleaned_words)

remove_at_word_udf = udf(remove_at_word, StringType())

df = df.withColumn('text', remove_at_word_udf(df['text']))

df.show(5)
```

```python
df.count()
```

### Plotting the Univariate Distribution of Sentiments

```python
sentiment_counts = df.groupBy('sentiment').count()
sentiment_counts_pd = sentiment_counts.toPandas()

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(sentiment_counts_pd['sentiment'], sentiment_counts_pd['count'], color='skyblue')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Distribution')

avg_sentimental_confidence = df.groupBy('sentiment').avg('sentiment_confidence')
sentiment_counts_pd = avg_sentimental_confidence.toPandas()

plt.subplot(1, 2, 2)
plt.bar(sentiment_counts_pd['sentiment'], sentiment_counts_pd['avg(sentiment_confidence)'], color='skyblue')
plt.xlabel('Sentiment')
plt.ylabel('Avg. Confidence')
plt.title('Sentiment Confidences')

plt.tight_layout()
plt.show()
```

### Text Cleaning

```python
characters = "[^\w\s]"
df = df.withColumn("text", regexp_replace(col("text"), characters, ""))
df = df.withColumn("text", lower(col("text")))
```

### Tokenization

```python
tokenizer = Tokenizer(inputCol="text", outputCol="words")
df = tokenizer.transform(df)
```

### StopWord Removal

```python
additional_stop_words = ['gopdebate', 'gopdebates', 'last', 'night', 'debate', 'candidate', 'trump', 'candidates']

stop_words = StopWordsRemover().getStopWords() + additional_stop_words

stop_words_remover = StopWordsRemover(inputCol="words", outputCol="filtered_words", stopWords=stop_words)
df = stop_words_remover.transform(df)
```

### Term Frequency

```python
hashingTF = HashingTF(inputCol="filtered_words", outputCol="rawFeatures", numFeatures=10000)
tf_df = hashingTF.transform(df)
```

### Inverse Document Frequency

```python
idf = IDF(inputCol="rawFeatures", outputCol="features")
idfModel = idf.fit(tf_df)
df = idfModel.transform(tf_df)
```

### Feature Selection

```python
selected_df = df.select("sentiment", "sentiment_confidence", "features")
```

```python
selected_df.show(5)
```

### Word Cloud

```python
def generate_wordcloud(df, sentiment):
    filtered_text = " ".join(df.filter(df.sentiment == sentiment).rdd.map(lambda x: " ".join(x.filtered_words)).collect())
    
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(filtered_text)
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'Word Cloud for {sentiment} Sentiment')
    plt.axis('off')
    plt.show()

sentiments = df.select("sentiment").distinct().rdd.map(lambda x: x[0]).collect()

for sentiment in sentiments:
    generate_wordcloud(df, sentiment)
```

### Label Encoder

```python
string_indexer = StringIndexer(inputCol='sentiment', outputCol='label')
df = string_indexer.fit(selected_df).transform(selected_df)
df.show(5)
```

```python
df.printSchema()
```

### Modelling Data

- Logistic Regression

```python
from pyspark.ml.classification import LogisticRegression


train_data, test_data = df.randomSplit([0.7, 0.3], seed=111)

logistic_reg = LogisticRegression(featuresCol='features', labelCol='label', maxIter=10, family='multinomial')

logistic_reg_model = logistic_reg.fit(train_data)

predictions = logistic_reg_model.transform(test_data)

evaluator = MulticlassClassificationEvaluator(predictionCol='prediction', labelCol='label', metricName='accuracy')
accuracy = evaluator.evaluate(predictions)

prediction_and_label = predictions.select('prediction', 'label').rdd.map(lambda x: (float(x[0]), float(x[1])))
metrics = MulticlassMetrics(prediction_and_label)
confusion_matrix = metrics.confusionMatrix()

labels = predictions.select('label').distinct().rdd.flatMap(lambda x: x).collect()

precision_recall_dict = {}
for label in labels:
    precision = metrics.precision(label)
    recall = metrics.recall(label)
    precision_recall_dict[label] = {'precision': precision, 'recall': recall}

for label, values in precision_recall_dict.items():
    print(f"Label: {label}")
    print(f"Precision: {values['precision']}")
    print(f"Recall: {values['recall']}")
    print()


print("Accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt
import numpy as np

confusion_matrix_data = confusion_matrix.toArray()

plt.figure(figsize=(8, 6))
plt.imshow(confusion_matrix_data, interpolation='nearest', cmap='Oranges')
plt.title('Confusion Matrix')
plt.colorbar()

tick_marks = np.arange(len(confusion_matrix_data))
plt.xticks(tick_marks, ['Negative', 'Neutral', 'Positive'])
plt.yticks(tick_marks, ['Negative', 'Neutral', 'Positive'])

for i in range(len(confusion_matrix_data)):
    for j in range(len(confusion_matrix_data[i])):
        plt.text(j, i, format(confusion_matrix_data[i][j], '.0f'),
                 horizontalalignment="center",
                 color="Black",
                 fontsize=12)

plt.xlabel('(Predicted)')
plt.ylabel('(Actual)')
plt.tight_layout()
plt.show()
```
